## 📦 Import Libraries

```python
import numpy as np
import pandas as pd
from numpy.linalg import norm, eig, det, inv
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

## 📁 Load Dataset
Replace this with your dataset file.

```python
# Example dataset
data = {
    'Math':[78,85,96,65,88],
    'Physics':[82,79,94,70,90],
    'Chemistry':[75,88,92,60,85]
}

df = pd.DataFrame(data)
df
```

# 🔹 Part A: Vector & Matrix Fundamentals

```python
# Represent each student as a vector
vectors = df.values
vectors
```

```python
# Norm-1 and Norm-2
norm1 = np.linalg.norm(vectors, ord=1)
norm2 = np.linalg.norm(vectors)

norm1, norm2
```

# 🔹 Part B: Matrix Operations

```python
matrix = df.values

transpose = matrix.T
determinant = det(matrix.T @ matrix)

transpose, determinant
```

# 🔹 Part C: Linear Transformations

In this dataset:

- Each subject represents a **dimension**.
- 2 subjects → 2D plane
- 3 subjects → 3D space
- Many subjects → **Higher dimensional space (hyperplane)**

# 🔹 Part D: Eigenvalues & Decomposition

```python
cov_matrix = np.cov(matrix, rowvar=False)
eigenvalues, eigenvectors = eig(cov_matrix)

eigenvalues, eigenvectors
```

# 🔹 Part E: PCA (Dimensionality Reduction)

```python
pca = PCA(n_components=2)
pca_result = pca.fit_transform(matrix)

pca_result
```

# 🔹 LDA Classification

```python
# Create simple labels
labels = ['Above','Above','Above','Below','Above']

lda = LDA(n_components=1)
lda_result = lda.fit_transform(matrix, labels)

lda_result
```
